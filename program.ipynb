```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('iris.csv')
plt.figure(figsize=(10,7))
x=data['sepal.length']
plt.hist(x,bins=20,color="r")
plt.title('sepal length')
plt.xlabel('sepal length')
plt.ylabel('count')
plt.show
```
